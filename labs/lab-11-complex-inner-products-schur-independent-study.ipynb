5. Schur decomposition and QR iteration;
6. normal matrices.


```python
import numpy as np
from numpy.linalg import norm, eig, qr
np.set_printoptions(precision=4, suppress=True)

def inner(u, v):
    """Convention: <u,v> = v^* u."""
    return np.vdot(v, u)

def is_hermitian(A):
    return np.allclose(A, A.conj().T)

def is_unitary(U):
    return np.allclose(U.conj().T @ U, np.eye(U.shape[0]))

def is_normal(A):
    return np.allclose(A.conj().T @ A, A @ A.conj().T)
```

## Question 1. Complex inner products

Let
\[
u=\begin{bmatrix}1+i\\2\end{bmatrix},
\qquad
v=\begin{bmatrix}i\\1-i\end{bmatrix}.
\]
Compute \(\langle u,v\rangle\), \(\langle v,u\rangle\), and \(\|u\|\).


```python
u = np.array([1+1j, 2+0j])
v = np.array([1j, 1-1j])
print("<u,v> =", inner(u, v))
print("<v,u> =", inner(v, u))
print("conjugate symmetry check:", np.allclose(inner(u,v), np.conj(inner(v,u))))
print("||u|| =", norm(u))
```

**Solution.**

\[
\langle u,v\rangle=3+i,
\qquad
\langle v,u\rangle=3-i,
\qquad
\|u\|=\sqrt6.
\]

### Similar practice

Let \(a=(2-i,1+i)\) and \(b=(1+i,-i)\). Compute \(\langle a,b\rangle\), \(\langle b,a\rangle\), and \(\|a\|\).


```python
a = np.array([2-1j, 1+1j])
b = np.array([1+1j, -1j])
print("<a,b> =", inner(a,b))
print("<b,a> =", inner(b,a))
print("||a|| =", norm(a))
```

## Question 2. Why conjugation matters

For \(z=(1,i)\), compare \(z^Tz\) and \(z^*z\).


```python
z = np.array([1+0j, 1j])
wrong = z.T @ z
right = z.conj().T @ z
print("z^T z =", wrong)
print("z^* z =", right)
```

**Solution.**

\[
z^Tz=1+i^2=0,
\]
which is not a valid length squared. But
\[
z^*z=|1|^2+|i|^2=2.
\]


## Question 3. Projection onto a complex line

Let
\[
u=\frac{1}{\sqrt2}(1,i),
\qquad
y=(2,1+i).
\]
Compute \(\operatorname{proj}_{\operatorname{span}(u)}(y)\).


```python
u = np.array([1, 1j]) / np.sqrt(2)
y = np.array([2+0j, 1+1j])
proj = inner(y, u) * u
residual = y - proj
print("<y,u> =", inner(y,u))
print("projection =", proj)
print("residual orthogonal to u?", np.allclose(inner(residual, u), 0))
```

**Solution.**

\[
\langle y,u\rangle=\frac{3-i}{\sqrt2},
\qquad
\operatorname{proj}(y)=\frac{3-i}{2}\begin{bmatrix}1\\i\end{bmatrix}.
\]


## Question 4. Hermitian and unitary checks

Check whether
\[
A=\begin{bmatrix}2&1+i\\1-i&3\end{bmatrix}
\]
is Hermitian and whether
\[
U=\frac1{\sqrt2}\begin{bmatrix}1&1\\1&-1\end{bmatrix}
\]
is unitary.


```python
A = np.array([[2, 1+1j], [1-1j, 3]], dtype=complex)
U = (1/np.sqrt(2))*np.array([[1,1],[1,-1]], dtype=complex)
print("A is Hermitian:", is_hermitian(A))
print("Eigenvalues of A:", np.linalg.eigvalsh(A))
print("U is unitary:", is_unitary(U))
print("U^* U =\n", U.conj().T @ U)
```

**Solution.**

\(A=A^*\), so \(A\) is Hermitian. Also \(U^*U=I\), so \(U\) is unitary.


## Question 5. Schur decomposition

Use SciPy, if available, to compute a Schur decomposition. If SciPy is not available, use NumPy eigenvalue computations and QR iteration as a substitute exploration.


```python
A = np.array([[13,8,8],[-1,7,-2],[-1,-2,7]], dtype=float)
try:
    from scipy.linalg import schur
    T, U = schur(A, output='complex')
    print("T =\n", T)
    print("Check A = U T U^*:", np.allclose(A, U @ T @ U.conj().T))
    print("Diagonal of T:", np.diag(T))
except Exception as e:
    print("SciPy Schur not available:", e)
    print("Eigenvalues from NumPy:", np.linalg.eigvals(A))
```

**Solution idea.**

Schur decomposition writes
\[
A=UTU^*,
\]
where \(U\) is unitary and \(T\) is upper triangular. The eigenvalues of \(A\) appear on the diagonal of \(T\).


## Question 6. QR iteration

Apply basic QR iteration to a real symmetric matrix and observe convergence toward diagonal form.


```python
A = np.array([[2,1,0],[1,2,1],[0,1,2]], dtype=float)
Ak = A.copy()
for k in range(25):
    Q, R = np.linalg.qr(Ak)
    Ak = R @ Q
print(Ak)
print("Approximate eigenvalues from diagonal:", np.diag(Ak))
print("Actual eigenvalues:", np.linalg.eigvalsh(A))
```

## Question 7. Normal but not Hermitian

Give a normal matrix that is not Hermitian and check it in Python.


```python
D = np.array([[1j,0],[0,1]], dtype=complex)
print("D is normal:", is_normal(D))
print("D is Hermitian:", is_hermitian(D))
print("D^*D =\n", D.conj().T @ D)
print("DD^* =\n", D @ D.conj().T)
```

**Solution.**

The diagonal matrix
\[
D=\begin{bmatrix}i&0\\0&1\end{bmatrix}
\]
is normal because diagonal matrices commute with their conjugate transposes. It is not Hermitian because \(D^*\ne D\).
