# Lab TDA: Topological Data Analysis from Linear Algebra

This independent-study notebook contains worked solutions and similar practice questions. The main goal is to compute Betti numbers from boundary matrices.


## 1. Basic functions

We use the formula

$$\beta_p=\dim C_p-\operatorname{rank}(\partial_p)-\operatorname{rank}(\partial_{p+1}).$$


```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations

def rank(A, tol=1e-10):
    A = np.array(A, dtype=float)
    if A.size == 0:
        return 0
    return np.linalg.matrix_rank(A, tol=tol)

def betti(dim_Cp, boundary_p, boundary_p_plus_1):
    return dim_Cp - rank(boundary_p) - rank(boundary_p_plus_1)
```

## 2. Worked example: triangle boundary

The triangle boundary has three vertices, three edges, and no filled face.


```python
D1 = np.array([
    [-1, -1,  0],
    [ 1,  0, -1],
    [ 0,  1,  1]
], dtype=float)
D2_empty = np.zeros((3, 0))
D0 = np.zeros((0, 3))

beta0 = betti(3, D0, D1)
beta1 = betti(3, D1, D2_empty)

print('rank(D1) =', rank(D1))
print('beta0 =', beta0)
print('beta1 =', beta1)
```

**Answer.** The triangle boundary is connected and has one loop, so $\beta_0=1$ and $\beta_1=1$.


## 3. Worked example: filled triangle

Now add the filled face $[1,2,3]$.


```python
D2_filled = np.array([[1], [-1], [1]], dtype=float)

beta0_filled = betti(3, D0, D1)
beta1_filled = betti(3, D1, D2_filled)
beta2_filled = betti(1, D2_filled, np.zeros((0, 1)))

print('rank(D2) =', rank(D2_filled))
print('beta0 =', beta0_filled)
print('beta1 =', beta1_filled)
print('beta2 =', beta2_filled)
```

**Answer.** Filling the triangle kills the loop. Therefore $\beta_1=0$.


## 4. Build boundary matrices automatically


```python
def simplex_faces(simplex):
    simplex = tuple(simplex)
    faces = []
    for i in range(len(simplex)):
        face = simplex[:i] + simplex[i+1:]
        sign = (-1)**i
        faces.append((face, sign))
    return faces

def boundary_matrix(p_simplices, pm1_simplices):
    row_index = {s:i for i,s in enumerate(pm1_simplices)}
    M = np.zeros((len(pm1_simplices), len(p_simplices)))
    for j, sigma in enumerate(p_simplices):
        for face, sign in simplex_faces(sigma):
            M[row_index[face], j] = sign
    return M
```

## 5. Task with solution: two triangles, one filled

Vertices: $1,2,3,4$

Edges: $[1,2],[1,3],[2,3],[2,4],[3,4]$

Face: $[2,3,4]$


```python
vertices = [(1,), (2,), (3,), (4,)]
edges = [(1,2), (1,3), (2,3), (2,4), (3,4)]
faces = [(2,3,4)]

D1_two = boundary_matrix(edges, vertices)
D2_two = boundary_matrix(faces, edges)
D0_two = np.zeros((0, len(vertices)))
D3_two = np.zeros((0, len(faces)))

print('D1 =')
print(D1_two.astype(int))
print('D2 =')
print(D2_two.astype(int))

print('beta0 =', betti(len(vertices), D0_two, D1_two))
print('beta1 =', betti(len(edges), D1_two, D2_two))
print('beta2 =', betti(len(faces), D2_two, D3_two))
```

**Answer.** The complex is connected and has one unfilled triangular loop. Thus $\beta_0=1$, $\beta_1=1$, and $\beta_2=0$.


## 6. Similar practice question: square cycle

Consider a square with vertices $1,2,3,4$ and edges $[1,2],[2,3],[3,4],[1,4]$, with no filled face. Compute $\beta_0$ and $\beta_1$.


```python
vertices_sq = [(1,), (2,), (3,), (4,)]
edges_sq = [(1,2), (2,3), (3,4), (1,4)]
faces_sq = []

D1_sq = boundary_matrix(edges_sq, vertices_sq)
D2_sq = np.zeros((len(edges_sq), 0))
D0_sq = np.zeros((0, len(vertices_sq)))

print(D1_sq.astype(int))
print('rank(D1) =', rank(D1_sq))
print('beta0 =', betti(len(vertices_sq), D0_sq, D1_sq))
print('beta1 =', betti(len(edges_sq), D1_sq, D2_sq))
```

**Answer.** The square is connected and has one loop, so $\beta_0=1$ and $\beta_1=1$.


## 7. Point cloud and Vietoris--Rips graph

This simple visualization builds only the graph layer. It shows how components merge as the scale increases.


```python
def pairwise_edges(X, r):
    edges = []
    for i, j in combinations(range(len(X)), 2):
        if np.linalg.norm(X[i] - X[j]) <= r:
            edges.append((i, j))
    return edges

def count_components(n, edges):
    parent = list(range(n))
    def find(x):
        while parent[x] != x:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x
    def union(a, b):
        a, b = find(a), find(b)
        if a != b:
            parent[b] = a
    for a, b in edges:
        union(a, b)
    return len({find(i) for i in range(n)})

theta = np.linspace(0, 2*np.pi, 16, endpoint=False)
X = np.c_[np.cos(theta), np.sin(theta)]

for r in [0.4, 0.8, 1.2, 1.8]:
    edges_r = pairwise_edges(X, r)
    comps = count_components(len(X), edges_r)
    cycle_rank = len(edges_r) - len(X) + comps
    print(f'r={r:.1f}: edges={len(edges_r)}, components={comps}, graph cycle rank={cycle_rank}')
```

```python
r = 0.8
edges_r = pairwise_edges(X, r)
plt.figure(figsize=(5,5))
for i, j in edges_r:
    plt.plot([X[i,0], X[j,0]], [X[i,1], X[j,1]], 'k-', alpha=0.4)
plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
plt.title(f'Vietoris--Rips graph at r={r}')
plt.show()
```
