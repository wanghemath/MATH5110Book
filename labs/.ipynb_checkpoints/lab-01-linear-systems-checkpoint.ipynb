## 1. Linear systems as questions

A **linear equation** in unknowns $x_1,\ldots,x_n$ has the form

$$
a_1x_1+a_2x_2+\cdots+a_nx_n=b,
$$

where $a_1,\ldots,a_n,b$ are scalars.

A **linear system** is a finite collection of linear equations. In matrix form, it is written

$$
Ax=b,
$$

where

$$
A\in \mathbb{R}^{m\times n},\qquad x\in \mathbb{R}^n,\qquad b\in \mathbb{R}^m.
$$

The **solution set** is

$$
\mathcal{S}=\{x\in \mathbb{R}^n: Ax=b\}.
$$

In this lab, we repeatedly ask: **What does the solution set look like?**

$$
\begin{aligned}

\end{aligned}
$$

In matrix form,

$$
A=\begin{bmatrix}1&1\\2&-1\end{bmatrix},
\qquad

$$

```python
A = sp.Matrix([[1, 1],
               [2, -1]])
b = sp.Matrix([4, 2])
A, b
```

```python
x1, x2 = sp.symbols('x1 x2')
solution = sp.solve(A * sp.Matrix([x1, x2]) - b, [x1, x2])
solution
```

## 3. Row reduction and the augmented matrix

The **augmented matrix** of $Ax=b$ is

$$
[A\mid b].
$$

Elementary row operations do not change the solution set of the system. They replace the original equations with equivalent equations.

The three elementary row operations are:

1. Swap two rows.
2. Multiply a row by a nonzero scalar.
3. Add a scalar multiple of one row to another row.

The reduced row-echelon form, or **RREF**, reveals the structure of the solution set.

```python
Aug = A.row_join(b)
Aug_rref, pivots = Aug.rref()
Aug, Aug_rref, pivots
```

## 4. The three possible endings

A linear system has exactly one of the following three outcomes:

1. **No solution**: the system is inconsistent.
2. **Exactly one solution**: the system is consistent and has no free variables.
3. **Infinitely many solutions**: the system is consistent and has at least one free variable.

```python
A_tmp = sp.Matrix([[1, 1], [2, -1]])
b_tmp = sp.Matrix([4, 2])
Aug_tmp = A_tmp.row_join(b_tmp)
Aug_tmp.rref()
```

### 4.x No solution

```python
A_tmp = sp.Matrix([[1, 1], [1, 1]])
b_tmp = sp.Matrix([1, 3])
Aug_tmp = A_tmp.row_join(b_tmp)
Aug_tmp.rref()
```

### 4.x Infinitely many solutions

```python
A_tmp = sp.Matrix([[1, 1, 1], [2, 2, 2]])
b_tmp = sp.Matrix([2, 4])
Aug_tmp = A_tmp.row_join(b_tmp)
Aug_tmp.rref()
```

A row of the form

$$
[0\quad 0\quad \cdots \quad 0\mid c], \qquad c\neq 0,
$$

means $0=c$, which is impossible. Therefore the system has no solution.

## 5. A rank test for consistency

For a system $Ax=b$:

- $\operatorname{rank}(A)$ is the number of pivot columns in $A$.
- $\operatorname{rank}([A\mid b])$ is the number of pivot columns in the augmented matrix.

The system $Ax=b$ is consistent if and only if

$$
\operatorname{rank}(A)=\operatorname{rank}([A\mid b]).
$$

If the system is consistent and $A\in\mathbb{R}^{m\times n}$, then the number of free variables is

$$
n-\operatorname{rank}(A).
$$

```python
def analyze_system(A, b):
    A = sp.Matrix(A)
    b = sp.Matrix(b)
    Aug = A.row_join(b)
    rank_A = A.rank()
    rank_Aug = Aug.rank()
    n = A.shape[1]
    consistent = (rank_A == rank_Aug)
    free_vars = n - rank_A if consistent else None
    return {
        "rank(A)": rank_A,
        "rank([A|b])": rank_Aug,
        "consistent?": consistent,
        "number of variables": n,
        "number of free variables": free_vars,
        "rref([A|b])": Aug.rref()[0]
    }
```

```python
A1 = sp.Matrix([[1, 1], [2, -1]])
b1 = sp.Matrix([4, 2])
A2 = sp.Matrix([[1, 1], [1, 1]])
b2 = sp.Matrix([1, 3])
A3 = sp.Matrix([[1, 1, 1], [2, 2, 2]])
b3 = sp.Matrix([2, 4])

analyze_system(A1, b1)
```

```python
analyze_system(A2, b2)
```

```python
analyze_system(A3, b3)
```

## 6. Geometry in two variables

A linear equation in two variables is a line. A system of two linear equations is an intersection problem for lines.

The three outcomes become: one intersection point, parallel distinct lines, or the same line.

```python
def plot_two_equations(eqs, xlim=(-5, 5), ylim=(-5, 5), title="Linear system"):
    x_vals = np.linspace(xlim[0], xlim[1], 400)
    plt.figure(figsize=(6, 6))
    for a, b, c, label in eqs:
        if abs(b) > 1e-12:
            y_vals = (c - a * x_vals) / b
            plt.plot(x_vals, y_vals, label=label)
        else:
            x0 = c / a
            plt.axvline(x0, label=label)
    plt.axhline(0, linewidth=0.8)
    plt.axvline(0, linewidth=0.8)
    plt.xlim(*xlim)
    plt.ylim(*ylim)
    plt.grid(True)
    plt.legend()
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

plot_two_equations([
    (1, 1, 4, r"$x+y=4$"),
    (2, -1, 2, r"$2x-y=2$")
], title="One solution")
```

```python
plot_two_equations([
    (1, 1, 1, r"$x+y=1$"),
    (1, 1, 3, r"$x+y=3$")
], title="No solution")
```

```python
plot_two_equations([
    (1, 1, 1, r"$x+y=1$"),
    (2, 2, 2, r"$2x+2y=2$")
], title="Infinitely many solutions")
```

## 7. Matrix equations as maps

The equation $Ax=b$ can be viewed as a question about a function

$$
T_A:\mathbb{R}^n\to \mathbb{R}^m,\qquad T_A(x)=Ax.
$$

The system $Ax=b$ asks whether $b$ is in the image of $T_A$.

Equivalently,

$$
Ax=b \text{ is consistent} \iff b\in \operatorname{Col}(A).
$$

To solve equations, understand the geometry of a linear map.

```python
A = sp.Matrix([[1, 2],
               [3, 6],
               [4, 8]])

b_in = sp.Matrix([1, 3, 4])
b_out = sp.Matrix([1, 0, 0])

print("Column space basis of A:")
display(A.columnspace())

print("Check b_in:")
display(analyze_system(A, b_in))

print("Check b_out:")
display(analyze_system(A, b_out))
```

## 8. Homogeneous systems and affine solution sets

The system $Ax=0$ is the **homogeneous system** associated with $Ax=b$.

If $x_p$ is one particular solution of $Ax=b$, then every solution has the form

$$
x=x_p+z,\qquad z\in \ker(A).
$$

Thus, when $Ax=b$ is consistent,

$$
\{x:Ax=b\}=x_p+\ker(A).
$$

A solution set is either empty, a single point, or an affine translate of the null space.

```python
A = sp.Matrix([[1, 1, 1],
               [2, 2, 2]])
b = sp.Matrix([2, 4])

print("RREF of augmented matrix:")
display(A.row_join(b).rref()[0])

print("Null space of A:")
display(A.nullspace())

x3 = sp.symbols('x3')
print("Solution set:")
sp.linsolve((A, b), x1, x2, x3)
```

## 9. Numerical computation versus exact computation

SymPy performs exact symbolic computation. NumPy performs numerical floating-point computation. For small exact examples, SymPy is often better. For large data problems, NumPy and SciPy are much faster.

```python
A_exact = sp.Matrix([[2, 1], [1, 3]])
b_exact = sp.Matrix([1, 4])

print("Exact solution:")
display(sp.linsolve((A_exact, b_exact)))

A_num = np.array([[2, 1], [1, 3]], dtype=float)
b_num = np.array([1, 4], dtype=float)

print("Numerical solution:")
print(np.linalg.solve(A_num, b_num))
```

Now try a nearly singular system.

```python
eps = 1e-10
A_near = np.array([[1, 1],
                   [1, 1 + eps]], dtype=float)
b_near = np.array([2, 2 + eps], dtype=float)

print("Condition number:", np.linalg.cond(A_near))
print("Numerical solution:", np.linalg.solve(A_near, b_near))
```

## 13. Optional challenge: automatic classification

Write a function that takes $A$ and $b$ and returns one of:

- `"no solution"`,
- `"unique solution"`,
- `"infinitely many solutions"`.

Use the rank theorem.

```python
def classify_system(A, b):
    A = sp.Matrix(A)
    b = sp.Matrix(b)
    Aug = A.row_join(b)
    rank_A = A.rank()
    rank_Aug = Aug.rank()
    n = A.shape[1]
    
    # TODO: complete the logic.
    if rank_A != rank_Aug:
        return "no solution"
    elif rank_A == n:
        return "unique solution"
    else:
        return "infinitely many solutions"

print(classify_system(A1, b1))
print(classify_system(A2, b2))
print(classify_system(A3, b3))
```
