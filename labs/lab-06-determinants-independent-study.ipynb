## 1. Determinant as area scaling

Let

$$A=\begin{bmatrix}2&1\\0&3\end{bmatrix}.$$

Compute $\det(A)$ and interpret it geometrically.


```python
A = sp.Matrix([[2, 1], [0, 3]])
detA = A.det()
detA
```

**Solution.** The determinant is $6$. Therefore $x\mapsto Ax$ maps the unit square to a parallelogram with area $6$. Since the determinant is positive, orientation is preserved.

**Similar practice.** Compute the determinant of $B=\begin{bmatrix}1&4\\0&1\end{bmatrix}$ and explain why this shear preserves area.


```python
B = sp.Matrix([[1, 4], [0, 1]])
B.det()
```

The determinant is $1$, so the shear preserves area.


## 2. Orientation

Compute the determinant of the reflection matrix

$$R=\begin{bmatrix}1&0\\0&-1\end{bmatrix}.$$


```python
R = sp.Matrix([[1, 0], [0, -1]])
R.det()
```

**Solution.** The determinant is $-1$. The transformation preserves area but reverses orientation.

**Similar practice.** A rotation by $90^\circ$ has matrix $Q=\begin{bmatrix}0&-1\\1&0\end{bmatrix}$. Compute $\det(Q)$.


```python
Q = sp.Matrix([[0, -1], [1, 0]])
Q.det()
```

The determinant is $1$, so this rotation preserves area and orientation.


## 3. Compute a 3 by 3 determinant by row operations

Let

$$A=\begin{bmatrix}1&2&3\\2&5&7\\1&1&0\end{bmatrix}.$$

Compute $\det(A)$.


```python
A = sp.Matrix([[1, 2, 3], [2, 5, 7], [1, 1, 0]])
A.det()
```

```python
# Row-replacement steps used in the notes
A1 = A.copy()
A1[1, :] = A1[1, :] - 2*A1[0, :]
A1[2, :] = A1[2, :] - A1[0, :]
A2 = A1.copy()
A2[2, :] = A2[2, :] + A2[1, :]
A1, A2, sp.prod(A2[i,i] for i in range(3))
```

**Solution.** The triangular form has diagonal entries $1,1,-2$. Since only row replacements were used, the determinant is $-2$.

**Similar practice.** Compute the determinant of

$$B=\begin{bmatrix}2&1&0\\1&3&1\\0&2&4\end{bmatrix}.$$


```python
B = sp.Matrix([[2, 1, 0], [1, 3, 1], [0, 2, 4]])
B.det()
```

The answer is $16$.


## 4. Determinant and invertibility

Let

$$A=\begin{bmatrix}1&2\\2&4\end{bmatrix}.$$

Determine whether $A$ is invertible.


```python
A = sp.Matrix([[1, 2], [2, 4]])
A.det(), A.rank(), A.rref()[0]
```

**Solution.** The determinant is $0$, so the matrix is not invertible. The rank is $1$, and the second column is twice the first column.

**Similar practice.** Determine whether $B=\begin{bmatrix}3&1\\2&5\end{bmatrix}$ is invertible.


```python
B = sp.Matrix([[3, 1], [2, 5]])
B.det(), B.inv()
```

The determinant is $13$, so $B$ is invertible.


## 5. Determinant from eigenvalues

Suppose a $4\times4$ matrix has eigenvalues $2,-1,3,5$. Find its determinant.


```python
eigs = [2, -1, 3, 5]
np.prod(eigs)
```

**Solution.** The determinant is $2(-1)(3)(5)=-30$, so the matrix is invertible.


## 6. Singular values and condition number

Suppose a $3\times3$ matrix has singular values $10,1,0.01$. Find $|\det(A)|$ and the condition number.


```python
singular_values = np.array([10.0, 1.0, 0.01])
abs_det = np.prod(singular_values)
cond = singular_values.max() / singular_values.min()
abs_det, cond
```

**Solution.** $|\det(A)|=0.1$ and $\kappa_2(A)=1000$. The matrix may be numerically unstable because one direction is much more compressed than another.


## 7. Covariance determinant and generalized variance

Let

$$\Sigma=\begin{bmatrix}4&0\\0&1\end{bmatrix}.$$

Compute $\det(\Sigma)$ and interpret it.


```python
Sigma = sp.Matrix([[4, 0], [0, 1]])
Sigma.det(), sp.sqrt(Sigma.det())
```

**Solution.** The generalized variance is $4$. The covariance ellipse has area proportional to $\sqrt{4}=2$.

**Similar practice.** Compute the generalized variance of $\begin{bmatrix}9&0\\0&4\end{bmatrix}$.


```python
Sigma2 = sp.Matrix([[9, 0], [0, 4]])
Sigma2.det(), sp.sqrt(Sigma2.det())
```

The generalized variance is $36$, and the ellipse area is proportional to $6$.


## 8. Log determinant with Cholesky

For a positive definite matrix, compute $\log\det(A)$ using Cholesky factorization.


```python
Sigma = np.array([[4.0, 1.2], [1.2, 2.0]])
L = np.linalg.cholesky(Sigma)
logdet_cholesky = 2*np.sum(np.log(np.diag(L)))
np.linalg.slogdet(Sigma), logdet_cholesky
```

**Solution.** The Cholesky method computes $\log\det(\Sigma)$ as $2\sum_i \log L_{ii}$, which is often more stable than computing the determinant first.
