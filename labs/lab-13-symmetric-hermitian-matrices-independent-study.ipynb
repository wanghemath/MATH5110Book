## 1. Symmetric and skew-symmetric decomposition

For any real square matrix $M$,

$$
M=\frac{M+M^T}{2}+\frac{M-M^T}{2}.
$$

The first term is symmetric and the second term is skew-symmetric.

```python
M = np.array([[1, 4],
              [2, 3]], dtype=float)
S = (M + M.T) / 2
K = (M - M.T) / 2
print("M =
", M)
print("S =
", S)
print("K =
", K)
print("S + K =
", S + K)
print("S symmetric?", np.allclose(S.T, S))
print("K skew-symmetric?", np.allclose(K.T, -K))
```

## 2. Hermitian matrices

A complex matrix $H$ is Hermitian if $H^*=H$, where $H^*$ is the conjugate transpose.

```python
H = np.array([[2, 1+1j],
              [1-1j, 4]], dtype=complex)
print("H =
", H)
print("H* =
", H.conj().T)
print("Hermitian?", np.allclose(H.conj().T, H))
```

### Similar practice

Check whether

$$
B=\begin{bmatrix}1&2+i\\2+i&3\end{bmatrix}
$$

is Hermitian.

```python
B = np.array([[1, 2+1j],
              [2+1j, 3]], dtype=complex)
print("B* =
", B.conj().T)
print("Hermitian?", np.allclose(B.conj().T, B))
```

## 3. Spectral decomposition

For symmetric or Hermitian matrices, use `np.linalg.eigh`.

```python
A = np.array([[3, 1],
              [1, 3]], dtype=float)
lam, Q = np.linalg.eigh(A)
D = np.diag(lam)
print("Eigenvalues:", lam)
print("Q =
", Q)
print("Q^T Q =
", Q.T @ Q)
print("Reconstruction error:", np.linalg.norm(A - Q @ D @ Q.T))
```

The columns of $Q$ are orthonormal eigenvectors. The matrix is reconstructed as

$$
A=QDQ^T.
$$

### Similar practice

Compute an orthogonal diagonalization of

$$
A=\begin{bmatrix}5&2\\2&5\end{bmatrix}.
$$

```python
A2 = np.array([[5, 2],
               [2, 5]], dtype=float)
lam2, Q2 = np.linalg.eigh(A2)
print("Eigenvalues:", lam2)
print("Q =
", Q2)
print("Reconstruction error:", np.linalg.norm(A2 - Q2 @ np.diag(lam2) @ Q2.T))
```

## 4. Quadratic forms and positive definiteness

A symmetric matrix $A$ defines

$$
q(x)=x^TAx.
$$

The matrix is positive definite if all eigenvalues are positive.

```python
def classify_symmetric(A, tol=1e-10):
    eigs = np.linalg.eigvalsh(A)
    if np.all(eigs > tol):
        cls = "positive definite"
    elif np.all(eigs >= -tol):
        cls = "positive semidefinite"
    elif np.all(eigs < -tol):
        cls = "negative definite"
    else:
        cls = "indefinite"
    return eigs, cls

for A in [np.array([[3,1],[1,3]], float),
          np.array([[1,1],[1,1]], float),
          np.array([[1,2],[2,1]], float)]:
    eigs, cls = classify_symmetric(A)
    print("A =
", A)
    print("eigenvalues:", eigs)
    print("classification:", cls)
    print()
```

### Similar practice

Classify

$$
C=\begin{bmatrix}2&0&0\\0&3&0\\0&0&-4\end{bmatrix}.
$$

```python
C = np.diag([2, 3, -4])
eigs, cls = classify_symmetric(C)
print("Eigenvalues:", eigs)
print("Classification:", cls)
```

## 5. Sylvester's criterion

A real symmetric matrix is positive definite if and only if all leading principal minors are positive.

```python
A = np.array([[2, 1, 0],
              [1, 3, 1],
              [0, 1, 2]], dtype=float)
mins = [np.linalg.det(A[:k,:k]) for k in range(1, A.shape[0]+1)]
print("Leading principal minors:", mins)
print("Positive definite by Sylvester?", all(m > 0 for m in mins))
print("Eigenvalue check:", np.linalg.eigvalsh(A))
```

### Similar practice

Apply Sylvester's criterion to

B=\begin{bmatrix}1&2\\2&1\end{bmatrix}.

```python
B = np.array([[1,2],[2,1]], dtype=float)
mins_B = [np.linalg.det(B[:k,:k]) for k in range(1, B.shape[0]+1)]
print("Leading principal minors:", mins_B)
print("Positive definite by Sylvester?", all(m > 0 for m in mins_B))
```

## 6. Gram matrices and $A^TA$

If the columns of $A$ are $v_1,\ldots,v_n$, then $A^TA$ is the Gram matrix of the columns.

```python
v1 = np.array([1, 2, 0], dtype=float)
v2 = np.array([0, 1, 3], dtype=float)
A = np.column_stack([v1, v2])
G = A.T @ A
print("A =
", A)
print("G = A^T A =
", G)
print("Eigenvalues of G:", np.linalg.eigvalsh(G))
print("Positive definite?", np.all(np.linalg.eigvalsh(G) > 1e-10))
```

### Dependent-vector example

If the columns are dependent, the Gram matrix is positive semidefinite but singular.

```python
v1 = np.array([1, 2, 3], dtype=float)
v2 = 2*v1
A = np.column_stack([v1, v2])
G = A.T @ A
print("G =
", G)
print("Eigenvalues:", np.linalg.eigvalsh(G))
print("det(G):", np.linalg.det(G))
```

## 7. Cholesky factorization

If $A$ is symmetric positive definite, then

$$
A=LL^T.
$$

```python
A = np.array([[4, 2, 0],
              [2, 5, 1],
              [0, 1, 3]], dtype=float)
L = np.linalg.cholesky(A)
print("L =
", L)
print("Reconstruction error:", np.linalg.norm(A - L @ L.T))
```
