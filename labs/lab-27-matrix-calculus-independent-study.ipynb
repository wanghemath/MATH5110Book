## 1. Setup

```python
import numpy as np
np.set_printoptions(precision=4, suppress=True)
```

## 2. Quadratic gradients

For

$$f(x)=\frac12 x^TQx-b^Tx,$$

where $Q=Q^T$, we have

$$\nabla f(x)=Qx-b,\qquad \nabla^2 f(x)=Q.$$

```python
Q = np.array([[4.0, 1.0], [1.0, 3.0]])
b = np.array([1.0, 2.0])
x = np.array([3.0, -2.0])

def f_quad(x):
    return 0.5 * x @ Q @ x - b @ x

def grad_quad(x):
    return Q @ x - b

print('f(x) =', f_quad(x))
print('gradient =', grad_quad(x))
print('Hessian =\n', Q)
print('critical point =', np.linalg.solve(Q, b))
```

## 3. Finite-difference gradient check

A centered finite difference estimates

$$\frac{\partial f}{\partial x_i}(x)\approx\frac{f(x+\varepsilon e_i)-f(x-\varepsilon e_i)}{2\varepsilon}.$$

```python
def finite_difference_grad(f, x, eps=1e-6):
    x = np.asarray(x, dtype=float)
    g = np.zeros_like(x)
    for i in range(len(x)):
        e = np.zeros_like(x)
        e[i] = 1.0
        g[i] = (f(x + eps*e) - f(x - eps*e))/(2*eps)
    return g

print('analytic:', grad_quad(x))
print('finite difference:', finite_difference_grad(f_quad, x))
```

## 4. Least squares gradient

For

$$f(x)=\frac12\|Ax-y\|_2^2,$$

we have

$$\nabla f(x)=A^T(Ax-y).$$

```python
A = np.array([[1.0, 1.0], [1.0, 2.0], [1.0, 3.0]])
y = np.array([1.0, 2.0, 2.0])
x = np.array([0.0, 0.0])

def f_ls(x):
    r = A @ x - y
    return 0.5 * r @ r

def grad_ls(x):
    return A.T @ (A @ x - y)

print('gradient at x:', grad_ls(x))
print('finite difference:', finite_difference_grad(f_ls, x))
print('least-squares solution:', np.linalg.lstsq(A, y, rcond=None)[0])
```

## 5. Gradient descent

```python
x = np.array([3.0, -2.0])
alpha = 0.15
history = [x.copy()]
for k in range(25):
    x = x - alpha * grad_quad(x)
    history.append(x.copy())
print('last iterate:', x)
print('exact minimizer:', np.linalg.solve(Q, b))
print('last objective:', f_quad(x))
```

## 6. Newton's method

For the quadratic objective, one Newton step solves the problem exactly when $Q$ is invertible.

```python
x0 = np.array([3.0, -2.0])
x_newton = x0 - np.linalg.solve(Q, grad_quad(x0))
print('Newton step:', x_newton)
print('exact minimizer:', np.linalg.solve(Q, b))
```

## 7. Matrix least squares

For

$$f(X)=\frac12\|MX-C\|_F^2,$$

we have

$$\nabla_X f(X)=M^T(MX-C).$$

```python
M = np.array([[1.0, 2.0], [0.0, 1.0]])
X = np.array([[1.0, 0.0], [0.0, 1.0]])
C = np.array([[2.0, 1.0], [0.0, 3.0]])

def f_mat(X):
    R = M @ X - C
    return 0.5 * np.sum(R*R)

def grad_mat(X):
    return M.T @ (M @ X - C)

print('f(X) =', f_mat(X))
print('gradient =\n', grad_mat(X))
```

## 8. Matrix gradient finite-difference check

```python
def finite_difference_matrix_grad(f, X, eps=1e-6):
    G = np.zeros_like(X, dtype=float)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            E = np.zeros_like(X, dtype=float)
            E[i, j] = 1.0
            G[i, j] = (f(X + eps*E) - f(X - eps*E))/(2*eps)
    return G

print('analytic matrix gradient:
', grad_mat(X))
print('finite-difference matrix gradient:
', finite_difference_matrix_grad(f_mat, X))
```

## 9. Two-sided matrix least squares

For

$$f(X)=\frac12\|AXB-C\|_F^2,$$

we have

$$\nabla_X f(X)=A^T(AXB-C)B^T.$$

```python
A2 = np.array([[1.0, 1.0], [0.0, 2.0]])
B2 = np.array([[2.0, 0.0], [1.0, 1.0]])
C2 = np.array([[1.0, 0.0], [0.0, 1.0]])
X2 = np.eye(2)

def f_two(X):
    R = A2 @ X @ B2 - C2
    return 0.5*np.sum(R*R)

def grad_two(X):
    return A2.T @ (A2 @ X @ B2 - C2) @ B2.T

print('analytic:
', grad_two(X2))
print('finite difference:
', finite_difference_matrix_grad(f_two, X2))
```
