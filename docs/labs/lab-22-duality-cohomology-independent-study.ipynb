## 1. Dual basis from an inverse matrix

If $\mathcal B=\{v_1,\ldots,v_n\}$ and $P=[v_1\ \cdots\ v_n]$, then the rows of $P^{-1}$ are the dual basis functionals.

### Worked example

Let

$$
v_1=(1,1),\qquad v_2=(1,2).
$$


```python
import numpy as np
from fractions import Fraction

P = np.array([[1, 1],
              [1, 2]], dtype=float)

Pinv = np.linalg.inv(P)

print("P =")
print(P)
print("\nP inverse =")
print(Pinv)
print("\nCheck P^{-1}P =")
print(Pinv @ P)
```

**Answer.** The dual basis is given by the rows of $P^{-1}$:

$$
\varphi^1(x,y)=2x-y,\qquad
\varphi^2(x,y)=-x+y.
$$


### Similar practice question

Let

$$
v_1=(1,2),\qquad v_2=(3,5).
$$

Compute the dual basis.


```python
P2 = np.array([[1, 3],
               [2, 5]], dtype=float)
Pinv2 = np.linalg.inv(P2)
print("P2 inverse =")
print(Pinv2)
print("\nCheck:")
print(Pinv2 @ P2)
```

**Solution.**

$$
P^{-1}=
\begin{bmatrix}
-5&3\\
2&-1
\end{bmatrix}.
$$

Thus

$$
\varphi^1(x,y)=-5x+3y,\qquad
\varphi^2(x,y)=2x-y.
$$


## 2. Coordinate extraction

If the rows of $P^{-1}$ are dual basis functionals, then

$$
[x]_{\mathcal B}=P^{-1}x.
$$

Use this to express $x=(4,7)$ in the basis $v_1=(1,1)$, $v_2=(1,2)$.


```python
x = np.array([4, 7], dtype=float)
coords = Pinv @ x
print("coordinates =", coords)
print("reconstruction =", P @ coords)
```

**Solution.**

The coordinate vector is

$$
[x]_{\mathcal B}=\begin{bmatrix}1\\3\end{bmatrix}.
$$

Indeed,

$$
(4,7)=1(1,1)+3(1,2).
$$


## 3. Annihilators

Let

$$
U=\operatorname{span}\{(1,1,0),(0,1,1)\}\subseteq\mathbb R^3.
$$

Find a basis for $U^\circ$.


```python
import sympy as sp

u1 = sp.Matrix([1, 1, 0])
u2 = sp.Matrix([0, 1, 1])

# Put spanning vectors as columns.
U = sp.Matrix.hstack(u1, u2)

print("U matrix:")
sp.pretty_print(U)

# A functional ell = [a b c] vanishes on U exactly when U.T * ell_col = 0.
ann_basis = U.T.nullspace()

print("\nBasis for annihilator:")
for v in ann_basis:
    sp.pretty_print(v)
```

**Solution.**

A basis for $U^\circ$ is represented by the row vector

$$
[-1,\ 1,\ -1].
$$

Thus one functional is

$$
\ell(x,y,z)=-x+y-z.
$$


### Similar practice question

Let

$$
U=\operatorname{span}\{(1,0,1),(1,1,0)\}.
$$

Find one nonzero functional in $U^\circ$.


```python
u1 = sp.Matrix([1, 0, 1])
u2 = sp.Matrix([1, 1, 0])
U2 = sp.Matrix.hstack(u1, u2)
ann_basis2 = U2.T.nullspace()
print("Basis for annihilator:")
for v in ann_basis2:
    sp.pretty_print(v)
```

**Solution.**

A basis is

$$
(1,-1,-1).
$$

So

$$
\ell(x,y,z)=x-y-z
$$

vanishes on both spanning vectors.


## 4. Dual maps and transposes

Let

$$
A=\begin{bmatrix}
1&2\\
2&4\\
1&1
\end{bmatrix}.
$$

Compute the matrix of the dual map.


```python
A = sp.Matrix([[1, 2],
               [2, 4],
               [1, 1]])
print("A =")
sp.pretty_print(A)
print("\nMatrix of dual map A.T =")
sp.pretty_print(A.T)
```

**Solution.**

The dual map is represented by

$$
A^T=
\begin{bmatrix}
1&2&1\\
2&4&1
\end{bmatrix}.
$$


## 5. Cohomology of a hollow triangle

For the hollow triangle, use

$$
D_1=
\begin{bmatrix}
-1&0&-1\\
1&-1&0\\
0&1&1
\end{bmatrix},
\qquad
D_2=0.
$$

Compute $\beta_0$ and $\beta_1$, and write the coboundary matrix $\delta^0$.


```python
D1 = sp.Matrix([
    [-1, 0, -1],
    [ 1,-1,  0],
    [ 0, 1,  1]
])

D2_hollow = sp.zeros(3, 0)

rank_D1 = D1.rank()
rank_D2 = D2_hollow.rank()

n0, n1, n2 = 3, 3, 0
beta0 = n0 - rank_D1
beta1 = n1 - rank_D1 - rank_D2

print("rank D1 =", rank_D1)
print("rank D2 =", rank_D2)
print("beta0 =", beta0)
print("beta1 =", beta1)

print("\ndelta^0 = D1.T")
sp.pretty_print(D1.T)
```

**Solution.**

The hollow triangle has

$$
\beta_0=1,\qquad \beta_1=1.
$$

There is one connected component and one loop.


## 6. Cohomology of a filled triangle

For the filled triangle,

$$
D_2=
\begin{bmatrix}
1\\
1\\
-1
\end{bmatrix}.
$$

Compute $\beta_0,\beta_1,\beta_2$ and the coboundary matrix $\delta^1$.


```python
D2_filled = sp.Matrix([[1], [1], [-1]])

rank_D1 = D1.rank()
rank_D2 = D2_filled.rank()

n0, n1, n2 = 3, 3, 1
beta0 = n0 - rank_D1
beta1 = n1 - rank_D1 - rank_D2
beta2 = n2 - rank_D2

print("rank D1 =", rank_D1)
print("rank D2 =", rank_D2)
print("beta0 =", beta0)
print("beta1 =", beta1)
print("beta2 =", beta2)

print("\ndelta^1 = D2.T")
sp.pretty_print(D2_filled.T)
```

**Solution.**

The filled triangle has

$$
\beta_0=1,\qquad \beta_1=0,\qquad \beta_2=0.
$$

The loop is filled by a $2$-simplex, so the first cohomology vanishes.
