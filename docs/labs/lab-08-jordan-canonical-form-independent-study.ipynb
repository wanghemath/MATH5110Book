## 0. Setup

```python
import sympy as sp
sp.init_printing()

def jordan_block(lam, k):
    J = lam*sp.eye(k)
    for i in range(k-1):
        J[i, i+1] = 1
    return J

def nilpotent_shift(k):
    N = sp.zeros(k)
    for i in range(k-1):
        N[i, i+1] = 1
    return N
```

## 1. A matrix that is not diagonalizable

Let

$$A=\begin{bmatrix}2&1\\0&2\end{bmatrix}.$$

We compute the eigenvalues and eigenspace.

```python
A = sp.Matrix([[2, 1], [0, 2]])
A.eigenvals(), (A - 2*sp.eye(2)).nullspace()
```

### Solution

The only eigenvalue is $2$, with algebraic multiplicity $2$. The eigenspace is one-dimensional:

$$E_2=\operatorname{span}\{(1,0)^T\}.$$

Therefore $A$ is not diagonalizable.

### Similar practice

Check whether $B=\begin{bmatrix}5&1\\0&5\end{bmatrix}$ is diagonalizable.

```python
B = sp.Matrix([[5, 1], [0, 5]])
B.eigenvals(), (B - 5*sp.eye(2)).nullspace()
```

### Answer

The eigenspace is one-dimensional while the algebraic multiplicity is two. Hence $B$ is not diagonalizable.

## 2. Jordan chains

For $A=\begin{bmatrix}2&1\\0&2\end{bmatrix}$, let $v_1=e_1$ and $v_2=e_2$.

```python
v1 = sp.Matrix([1, 0])
v2 = sp.Matrix([0, 1])
N = A - 2*sp.eye(2)
N*v1, N*v2
```

### Solution

The output shows

$$Nv_1=0,\qquad Nv_2=v_1.$$

Thus $v_1,v_2$ is a Jordan chain.

## 3. Powers of a Jordan block

For $J=J_{\lambda,k}=\lambda I+N$,

$$J^m=\sum_{r=0}^{k-1}\binom{m}{r}\lambda^{m-r}N^r.$$

```python
lam = sp.Symbol('lambda')
m = sp.Symbol('m', integer=True, nonnegative=True)
J = jordan_block(3, 3)
J
```

```python
J**4
```

### Worked formula

For

$$J=\begin{bmatrix}3&1&0\\0&3&1\\0&0&3\end{bmatrix},$$

we have

$$J^m=3^mI+m3^{m-1}N+\binom{m}{2}3^{m-2}N^2.$$

```python
mval = 4
N3 = nilpotent_shift(3)
formula = 3**mval*sp.eye(3) + mval*3**(mval-1)*N3 + sp.binomial(mval, 2)*3**(mval-2)*(N3**2)
formula, J**mval, sp.simplify(formula - J**mval)
```

### Answer

## 4. Matrix exponentials

For $J=\lambda I+N$,

$$e^{tJ}=e^{\lambda t}\sum_{r=0}^{k-1}\frac{t^r}{r!}N^r.$$

```python
t = sp.Symbol('t')
J2 = jordan_block(2, 2)
sp.exp(t*J2)
```

### Solution

SymPy confirms

$$e^{tJ}=e^{2t}\begin{bmatrix}1&t\\0&1\end{bmatrix}.$$

## 5. Kernel dimensions and Jordan block sizes

If $d_j=\dim\ker((A-\lambda I)^j)$, then $d_j-d_{j-1}$ counts the number of Jordan blocks of size at least $j$.

```python
def block_sizes_from_kernel_dims(ds):
    # ds should be [d1, d2, ..., dr]
    ds = [0] + list(ds)
    diffs = [ds[i] - ds[i-1] for i in range(1, len(ds))]
    sizes = []
    for j in range(1, len(diffs)):
        exact = diffs[j-1] - diffs[j]
        sizes += [j]*exact
    sizes += [len(diffs)]*diffs[-1]
    return sorted(sizes, reverse=True), diffs

block_sizes_from_kernel_dims([2,4,5,5])
```

### Solution

The block sizes are $3$ and $2$.

### Similar practice

Use $d_1=3,d_2=5,d_3=6,d_4=6$.

```python
block_sizes_from_kernel_dims([3,5,6,6])
```

### Answer

The block sizes are $3,2,1$.

## 6. Minimal polynomial

The minimal polynomial records the largest Jordan block size for each eigenvalue.

```python
x = sp.Symbol('x')
minimal_polynomial_example = (x-1)**3 * (x+2)
sp.expand(minimal_polynomial_example)
```

### Solution

If the largest Jordan block for eigenvalue $1$ has size $3$ and the largest Jordan block for eigenvalue $-2$ has size $1$, then

$$m_A(x)=(x-1)^3(x+2).$$
