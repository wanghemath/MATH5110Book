## 0. Setup

```
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=6, suppress=True)
```

## 1. Least squares as projection

We use the lecture example. The residual should be orthogonal to the column space of $A$.

```
A = np.array([[-1, 4], [1, 8], [-1, 4]], dtype=float)
b = np.array([14, -4, 0], dtype=float)

x_star = np.linalg.lstsq(A, b, rcond=None)[0]
fitted = A @ x_star
r = b - fitted

print('x* =', x_star)
print('A x* =', fitted)
print('residual =', r)
print('||r|| =', np.linalg.norm(r))
print('A.T @ r =', A.T @ r)
```

**Similar practice.** Change $b$ to another vector and verify that $A^T r=0$ for the least squares residual.

```
b2 = np.array([2, 1, 5], dtype=float)
x2 = np.linalg.lstsq(A, b2, rcond=None)[0]
r2 = b2 - A @ x2
print('x2 =', x2)
print('A.T @ r2 =', A.T @ r2)
```

## 2. Normal equations by hand and by Python

```
A = np.array([[1,0],[1,1],[1,2]], dtype=float)
b = np.array([1,2,4], dtype=float)

normal_matrix = A.T @ A
rhs = A.T @ b
c = np.linalg.solve(normal_matrix, rhs)

print('A.T A =
', normal_matrix)
print('A.T b =', rhs)
print('coefficients [c0,c1] =', c)
```

The least squares line is $h(t)=c_0+c_1t$.

```
t = np.array([0,1,2], dtype=float)
y = b
grid = np.linspace(-0.2,2.2,100)
plt.scatter(t,y,label='data')
plt.plot(grid, c[0] + c[1]*grid, label='least squares line')
plt.legend(); plt.xlabel('t'); plt.ylabel('y'); plt.show()
```

## 3. Polynomial fitting

```
t = np.array([0, 1, -1, 2], dtype=float)
y = np.array([5, 3, 13, 1], dtype=float)

def design(t, degree):
    return np.vander(t, N=degree+1, increasing=True)

for degree in [1, 2, 3]:
    A = design(t, degree)
    coeff = np.linalg.lstsq(A, y, rcond=None)[0]
    residual = y - A @ coeff
    print('degree', degree)
    print('coefficients:', coeff)
    print('residual norm:', np.linalg.norm(residual))
    print()
```

```
grid = np.linspace(-1.3, 2.3, 300)
plt.scatter(t, y, label='data')
for degree in [1, 2, 3]:
    A = design(t, degree)
    coeff = np.linalg.lstsq(A, y, rcond=None)[0]
    G = design(grid, degree)
    plt.plot(grid, G @ coeff, label=f'degree {degree}')
plt.legend(); plt.xlabel('t'); plt.ylabel('y'); plt.title('Polynomial least squares fits'); plt.show()
```

## 4. Weighted least squares

```
t = np.array([0,1,2,3,4], dtype=float)
y = np.array([1,2,3,20,5], dtype=float)
A = design(t, 1)

# Ordinary least squares
c_ols = np.linalg.lstsq(A, y, rcond=None)[0]

# Weighted least squares: downweight the outlier at t=3
w = np.array([1,1,1,0.05,1], dtype=float)
W = np.diag(w)
c_wls = np.linalg.solve(A.T @ W @ A, A.T @ W @ y)

print('ordinary coefficients:', c_ols)
print('weighted coefficients:', c_wls)

grid = np.linspace(-0.2,4.2,200)
plt.scatter(t,y,label='data')
plt.plot(grid, c_ols[0]+c_ols[1]*grid, label='ordinary LS')
plt.plot(grid, c_wls[0]+c_wls[1]*grid, label='weighted LS')
plt.legend(); plt.show()
```

## 5. QR method

```
A = np.array([[1,0],[1,1],[1,2],[1,3]], dtype=float)
b = np.array([1,2,2.9,4.2], dtype=float)

Q, R = np.linalg.qr(A)
x_qr = np.linalg.solve(R, Q.T @ b)
x_np = np.linalg.lstsq(A, b, rcond=None)[0]
print('QR solution:', x_qr)
print('np.linalg.lstsq solution:', x_np)
```

## 6. Best function approximation

```
# Best linear approximation of exp(x) on [0,1]
e = np.e
G = np.array([[1, 1/2], [1/2, 1/3]], dtype=float)
rhs = np.array([e-1, 1], dtype=float)
c = np.linalg.solve(G, rhs)
print('coefficients:', c)
print('formula: p(x) = (4e-10) + (18-6e)x')

x = np.linspace(0,1,200)
plt.plot(x, np.exp(x), label='exp(x)')
plt.plot(x, c[0]+c[1]*x, label='best linear approximation')
plt.legend(); plt.show()
```
