## 1. Outer products and simple tensors

For $u\in\mathbb R^m$ and $v\in\mathbb R^n$, the tensor $u\otimes v$ corresponds to the matrix $uv^T$.

```python
u = np.array([1, 2, 3])
v = np.array([4, 5])
T = np.outer(u, v)
print(T)
print("rank =", np.linalg.matrix_rank(T))
```

### Worked solution

The result is

$$
\begin{bmatrix}4&5\\8&10\\12&15\end{bmatrix}.
$$

It has rank $1$ because every column is a scalar multiple of $u$.

### Similar practice

Compute $a\otimes b$ for $a=(2,-1)^T$ and $b=(3,0,4)^T$.

```python
a = np.array([2, -1])
b = np.array([3, 0, 4])
print(np.outer(a, b))
```

## 2. Kronecker products

The Kronecker product $A\otimes B$ is the block matrix $[a_{ij}B]$.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[0, 5], [6, 7]])
print(np.kron(A, B))
```

### Similar practice

Compute

$$
\begin{bmatrix}1&0\\0&-1\end{bmatrix}\otimes
\begin{bmatrix}2&1\\1&2\end{bmatrix}.
$$

```python
A = np.array([[1, 0], [0, -1]])
B = np.array([[2, 1], [1, 2]])
print(np.kron(A, B))
```

## 3. Vectorization identity

We verify

$$
\operatorname{vec}(AXB)=(B^T\otimes A)\operatorname{vec}(X).
$$

Use column stacking: `reshape(-1, order="F")`.

```python
A = np.array([[1, 2], [0, 1]])
X = np.array([[1, 3], [2, 4]])
B = np.array([[2, 0], [1, 1]])

left = (A @ X @ B).reshape(-1, order="F")
right = np.kron(B.T, A) @ X.reshape(-1, order="F")
print("left =", left)
print("right =", right)
print("verified:", np.allclose(left, right))
```

### Similar practice

Choose your own small matrices and verify the same identity.

```python
A = np.array([[2, 1], [1, 0]])
X = np.array([[1, 2], [3, 1]])
B = np.array([[0, 1], [2, 1]])
left = (A @ X @ B).reshape(-1, order="F")
right = np.kron(B.T, A) @ X.reshape(-1, order="F")
print(left)
print(right)
print(np.allclose(left, right))
```

## 4. Wedge products and signed area

For $u=(a,b)$ and $v=(c,d)$,

$$
u\wedge v=(ad-bc)e_1\wedge e_2.
$$

```python
u = np.array([2, 1])
v = np.array([1, 3])
signed_area = np.linalg.det(np.column_stack([u, v]))
print("signed area =", signed_area)
print("ordinary area =", abs(signed_area))
```

### Similar practice

Compute the signed area for $a=(1,4)$ and $b=(3,2)$.

```python
a = np.array([1, 4])
b = np.array([3, 2])
print(np.linalg.det(np.column_stack([a, b])))
```

## 5. Third-order rank-one tensors

An order-3 rank-one tensor has entries

$$
T_{ijk}=a_i b_j c_k.
$$

```python
a = np.array([1, 2])
b = np.array([3, 4, 5])
c = np.array([6, 7])
T = np.einsum('i,j,k->ijk', a, b, c)
print("shape:", T.shape)
print("slice k=0")
print(T[:, :, 0])
print("slice k=1")
print(T[:, :, 1])
```

## 6. Synthetic tensor and SVD of unfoldings

Construct a third-order tensor as a sum of two rank-one tensors. Then inspect the singular values of one unfolding.

```python
rng = np.random.default_rng(1)
a1, a2 = rng.normal(size=4), rng.normal(size=4)
b1, b2 = rng.normal(size=5), rng.normal(size=5)
c1, c2 = rng.normal(size=3), rng.normal(size=3)

T = 2*np.einsum('i,j,k->ijk', a1,b1,c1) - np.einsum('i,j,k->ijk', a2,b2,c2)
T1 = T.reshape(4, 15)
_, s, _ = np.linalg.svd(T1, full_matrices=False)
print(s)
```

## 7. Two-qubit separability

A two-qubit state

$$
\alpha|00\rangle+\beta|01\rangle+\gamma|10\rangle+\delta|11\rangle
$$

is separable if and only if

$$
\alpha\delta-\beta\gamma=0.
$$

```python
def separability(alpha, beta, gamma, delta):
    M = np.array([[alpha, beta], [gamma, delta]], dtype=float)
    return M, np.linalg.det(M), np.linalg.matrix_rank(M)

M, detM, rankM = separability(1/np.sqrt(2), 0, 0, 1/np.sqrt(2))
print(M)
print("det =", detM)
print("rank =", rankM)
print("separable?", abs(detM) < 1e-10)
```

### Similar practice

Test

$$
\phi=\frac12(|00\rangle+|01\rangle+|10\rangle+|11\rangle).
$$

```python
M, detM, rankM = separability(0.5, 0.5, 0.5, 0.5)
print(M)
print("det =", detM)
print("rank =", rankM)
print("separable?", abs(detM) < 1e-10)
```
