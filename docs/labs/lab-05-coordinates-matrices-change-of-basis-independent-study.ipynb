Main ideas:

1. A basis is a coordinate system.
2. Coordinates depend on the basis.
3. A linear transformation becomes a matrix only after choosing bases.
4. Change of basis translates one coordinate language into another.
5. Similar matrices represent the same transformation written in different bases.

```python
import sympy as sp
sp.init_printing()
```

## 1. Coordinates in a basis

Let

$$
b_1=\begin{bmatrix}1\\1\end{bmatrix},\qquad
b_2=\begin{bmatrix}-1\\2\end{bmatrix},\qquad
x=\begin{bmatrix}2\\8\end{bmatrix}.
$$

Find $[x]_{\mathcal B}$ where $\mathcal B=(b_1,b_2)$.

### Solution

Build the basis matrix $P_{\mathcal B}=[b_1\ b_2]$ and solve

$$
P_{\mathcal B}[x]_{\mathcal B}=x.
$$

```python
b1 = sp.Matrix([1, 1])
b2 = sp.Matrix([-1, 2])
P_B = sp.Matrix.hstack(b1, b2)
x = sp.Matrix([2, 8])
coords_B = P_B.LUsolve(x)
P_B, coords_B, P_B*coords_B
```

Therefore

$$
[x]_{\mathcal B}=\begin{bmatrix}4\\2\end{bmatrix}.
$$

### Similar practice

Find $[y]_{\mathcal B}$ for $y=(5,7)^T$.

**Answer:** $[y]_{\mathcal B}=(17/3,2/3)^T$.

```python
y = sp.Matrix([5, 7])
P_B.LUsolve(y)
```

## 2. Reconstructing a vector from coordinates

Suppose

$$
[v]_{\mathcal B}=\begin{bmatrix}3\\-1\end{bmatrix}.
$$

Find the actual vector $v$ in standard coordinates.

### Solution

Use

$$
v=P_{\mathcal B}[v]_{\mathcal B}.
$$

```python
coord_v_B = sp.Matrix([3, -1])
v = P_B * coord_v_B
v
```

Thus

$$
v=\begin{bmatrix}4\\1\end{bmatrix}.
$$

The coordinate vector and the vector are different objects. The coordinate vector tells us how to build the vector from the basis.

### Similar practice

If $[w]_{\mathcal B}=(2,5)^T$, find $w$.

**Answer:** $w=(-3,12)^T$.

```python
P_B * sp.Matrix([2, 5])
```

## 3. Matrix of a linear transformation in chosen bases

Let $T:\mathbb R^2\to\mathbb R^2$ be given by

$$
T(x)=Ax,\qquad
A=\begin{bmatrix}2&1\\0&3\end{bmatrix}.
$$

Let $\mathcal B$ be the basis above and let $\mathcal C$ be the standard basis. Find $[T]_{\mathcal C\leftarrow\mathcal B}$.

### Solution

The formula is

$$
[T]_{\mathcal C\leftarrow\mathcal B}=P_{\mathcal C}^{-1}AP_{\mathcal B}.
$$

Since $\mathcal C$ is standard, $P_{\mathcal C}=I$.

```python
A = sp.Matrix([[2, 1], [0, 3]])
P_C_standard = sp.eye(2)
T_C_from_B = P_C_standard.inv() * A * P_B
T_C_from_B
```

So

$$
[T]_{\mathcal C\leftarrow\mathcal B}
=\begin{bmatrix}3&0\\3&6\end{bmatrix}.
$$

The columns are $T(b_1)$ and $T(b_2)$ written in $\mathcal C$-coordinates.

### Similar practice

Let

$$
P_{\mathcal C}=\begin{bmatrix}1&1\\1&-1\end{bmatrix}.
$$

Compute $[T]_{\mathcal C\leftarrow\mathcal B}$.

**Answer:**

$$
\begin{bmatrix}3&3\\0&-3\end{bmatrix}.
$$

```python
P_C = sp.Matrix([[1, 1], [1, -1]])
P_C.inv() * A * P_B
```

## 4. Change of coordinates

Suppose

$$
[v]_{\mathcal B}=\begin{bmatrix}4\\2\end{bmatrix}.
$$

Find $[v]_{\mathcal C}$ where

$$
P_{\mathcal C}=\begin{bmatrix}1&1\\1&-1\end{bmatrix}.
$$

### Solution

Use

$$
[v]_{\mathcal C}=P_{\mathcal C}^{-1}P_{\mathcal B}[v]_{\mathcal B}.
$$

```python
coord_v_B = sp.Matrix([4, 2])
change_C_from_B = P_C.inv() * P_B
coord_v_C = change_C_from_B * coord_v_B
change_C_from_B, coord_v_C
```

Thus

$$
[v]_{\mathcal C}=\begin{bmatrix}5\\-3\end{bmatrix}.
$$

### Similar practice

Convert $[w]_{\mathcal B}=(1,3)^T$ into $\mathcal C$-coordinates.

**Answer:** $[w]_{\mathcal C}=(5/2,-9/2)^T$.

```python
change_C_from_B * sp.Matrix([1, 3])
```

## 5. Similarity and diagonalization

Let

$$
A=\begin{bmatrix}2&1\\0&3\end{bmatrix},
\qquad
P=\begin{bmatrix}1&1\\0&1\end{bmatrix}.
$$

Compute $P^{-1}AP$.

### Solution

When the same basis is used for the domain and codomain, the new matrix is

$$
A_{\mathcal B}=P^{-1}AP.
$$

```python
P = sp.Matrix([[1, 1], [0, 1]])
A_B = P.inv() * A * P
A_B
```

This gives

$$
A_{\mathcal B}=\begin{bmatrix}2&0\\0&3\end{bmatrix}.
$$

So this basis is an eigenbasis. In this coordinate system, the map only scales coordinate directions.

### Check invariants

Similar matrices have the same determinant, trace, and eigenvalues.

```python
A.det(), A_B.det(), A.trace(), A_B.trace(), A.eigenvals(), A_B.eigenvals()
```

### Similar practice

Let

$$
A=\begin{bmatrix}4&2\\1&3\end{bmatrix},
\qquad
P=\begin{bmatrix}2&1\\1&-1\end{bmatrix}.
$$

Compute $P^{-1}AP$.

**Answer:** $\operatorname{diag}(5,2)$.

```python
A2 = sp.Matrix([[4, 2], [1, 3]])
P2 = sp.Matrix([[2, 1], [1, -1]])
P2.inv() * A2 * P2
```

## 6. Challenge question

Let $T:P_2\to P_2$ be the derivative-like transformation

$$
T(p)=p+p'.
$$

Use the standard polynomial basis $\mathcal E=(1,t,t^2)$. Find the matrix of $T$ in this basis.

### Solution

Compute the images of the basis vectors:

$$
T(1)=1,
$$

$$
T(t)=t+1,
$$

$$
T(t^2)=t^2+2t.
$$

The coordinate columns in the basis $(1,t,t^2)$ are

$$
[1]_{\mathcal E}=\begin{bmatrix}1\\0\\0\end{bmatrix},\quad
[1+t]_{\mathcal E}=\begin{bmatrix}1\\1\\0\end{bmatrix},\quad
[2t+t^2]_{\mathcal E}=\begin{bmatrix}0\\2\\1\end{bmatrix}.
$$

Therefore

$$
[T]_{\mathcal E}=\begin{bmatrix}
1&1&0\\
0&1&2\\
0&0&1
\end{bmatrix}.
$$

```python
T_poly = sp.Matrix([[1, 1, 0],
                    [0, 1, 2],
                    [0, 0, 1]])
T_poly
```
