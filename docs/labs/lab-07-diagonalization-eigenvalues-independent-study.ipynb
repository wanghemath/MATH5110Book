Main ideas:

- An eigenvector is a nonzero vector $v$ such that $Av=\lambda v$.
- The eigenspace for $\lambda$ is $\ker(A-\lambda I)$.
- A matrix is diagonalizable when it has enough linearly independent eigenvectors.
- If $A=PDP^{-1}$, then $A^k=PD^kP^{-1}$.
- Markov steady states are eigenvectors with eigenvalue $1$.


```
import numpy as np
import sympy as sp
np.set_printoptions(precision=4, suppress=True)

x = sp.symbols('x')

def rref_np(A):
    return sp.Matrix(A).rref()

def eig_summary(A):
    M = sp.Matrix(A)
    print("A =")
    display(M)
    print("Characteristic polynomial:")
    display(M.charpoly(x).as_expr().factor())
    print("Eigenvects:")
    for lam, alg_mult, basis in M.eigenvects():
        print(f"lambda = {lam}, algebraic multiplicity = {alg_mult}, geometric multiplicity = {len(basis)}")
        for v in basis:
            display(v)
```

## 1. Check an eigenvector

Let

$$
A=\begin{bmatrix}2&1\\0&3\end{bmatrix},\qquad v=\begin{bmatrix}1\\0\end{bmatrix}.
$$

Check whether $v$ is an eigenvector.


```
A = sp.Matrix([[2, 1], [0, 3]])
v = sp.Matrix([1, 0])
Av = A*v
Av
```

Since

$$
Av=\begin{bmatrix}2\\0\end{bmatrix}=2v,
$$

$v$ is an eigenvector with eigenvalue $2$.

### Similar practice

Check whether $w=(1,1)^T$ is an eigenvector for the same matrix.


```
w = sp.Matrix([1, 1])
Aw = A*w
Aw, 3*w, Aw == 3*w
```

**Answer.** Yes. Since $Aw=3w$, the vector $w$ is an eigenvector with eigenvalue $3$.

## 2. Eigenvalues and eigenspaces

Let

$$
A=\begin{bmatrix}4&1\\2&3\end{bmatrix}.
$$

Find the eigenvalues and eigenspaces.


```
A = sp.Matrix([[4, 1], [2, 3]])
eig_summary(A)
```

The characteristic polynomial is

$$
\lambda^2-7\lambda+10=(\lambda-5)(\lambda-2).
$$

Thus the eigenvalues are $5$ and $2$. Corresponding eigenvectors are $(1,1)^T$ and $(1,-2)^T$.

### Similar practice

Find the eigenvalues and eigenspaces of

$$
B=\begin{bmatrix}3&1\\0&2\end{bmatrix}.
$$


```
B = sp.Matrix([[3, 1], [0, 2]])
eig_summary(B)
```

**Answer.** The eigenvalues are $3$ and $2$. One eigenbasis is $(1,0)^T$ for $\lambda=3$ and $(-1,1)^T$ for $\lambda=2$.

## 3. Diagonalization

Diagonalize

$$
A=\begin{bmatrix}4&1\\2&3\end{bmatrix}.
$$


```
A = sp.Matrix([[4, 1], [2, 3]])
P = sp.Matrix([[1, 1], [1, -2]])
D = sp.diag(5, 2)
print("P =")
display(P)
print("D =")
display(D)
print("P D P^{-1} =")
display(P*D*P.inv())
print("Check P D P^{-1} == A:", P*D*P.inv() == A)
```

The columns of $P$ are eigenvectors of $A$, and the diagonal entries of $D$ are the corresponding eigenvalues. Therefore

$$
A=PDP^{-1}.
$$

### Similar practice

Diagonalize $B=\begin{bmatrix}3&1\\0&2\end{bmatrix}$.


```
B = sp.Matrix([[3, 1], [0, 2]])
P_B = sp.Matrix([[1, -1], [0, 1]])
D_B = sp.diag(3, 2)
display(P_B*D_B*P_B.inv())
print(P_B*D_B*P_B.inv() == B)
```

**Answer.** One diagonalization is

$$
B=\begin{bmatrix}1&-1\\0&1\end{bmatrix}
\begin{bmatrix}3&0\\0&2\end{bmatrix}
\begin{bmatrix}1&-1\\0&1\end{bmatrix}^{-1}.
$$

## 4. Matrix powers

Use diagonalization to compute $A^{10}$ for

$$
A=\begin{bmatrix}4&1\\2&3\end{bmatrix}.
$$


```
k = 10
A10_by_diag = P*(D**k)*P.inv()
A10_direct = A**k
display(A10_by_diag)
print("Matches direct power:", A10_by_diag == A10_direct)
```

The formula is

$$
A^{10}=P\begin{bmatrix}5^{10}&0\\0&2^{10}\end{bmatrix}P^{-1}.
$$

### Similar practice

Use the same idea to compute $B^8$ for $B=\begin{bmatrix}3&1\\0&2\end{bmatrix}$.


```
k = 8
B8_by_diag = P_B*(D_B**k)*P_B.inv()
display(B8_by_diag)
print("Matches direct power:", B8_by_diag == B**k)
```

## 5. A matrix that is not diagonalizable

Let

$$
J=\begin{bmatrix}2&1\\0&2\end{bmatrix}.
$$

Show that $J$ is not diagonalizable.


```
J = sp.Matrix([[2, 1], [0, 2]])
eig_summary(J)
```

The only eigenvalue is $2$, and the eigenspace has dimension $1$. A $2\times2$ diagonalizable matrix needs two linearly independent eigenvectors. Therefore $J$ is not diagonalizable.

### Similar practice

Is $C=\begin{bmatrix}4&0\\0&4\end{bmatrix}$ diagonalizable?


```
C = sp.Matrix([[4, 0], [0, 4]])
eig_summary(C)
```

**Answer.** Yes. The eigenspace for $\lambda=4$ is all of $\mathbb R^2$, so there are two linearly independent eigenvectors.

## 6. Markov-chain steady state

Let

$$
M=\begin{bmatrix}0.8&0.3\\0.2&0.7\end{bmatrix}.
$$

Find the steady state $p$ satisfying $Mp=p$ and $p_1+p_2=1$.


```
M = sp.Matrix([[sp.Rational(8,10), sp.Rational(3,10)], [sp.Rational(2,10), sp.Rational(7,10)]])
p1, p2 = sp.symbols('p1 p2')
sol = sp.solve([sp.Eq((M*sp.Matrix([p1,p2]))[0], p1), sp.Eq(p1+p2, 1)], [p1,p2])
sol
```

Thus

$$
p=\begin{bmatrix}0.6\\0.4\end{bmatrix}.
$$

We can also see convergence by applying $M$ many times to an initial probability vector.

```
p = sp.Matrix([1, 0])
for k in range(10):
    p = M*p
p.evalf()
```

### Similar practice

Find the steady state of

N=\begin{bmatrix}0.9&0.4\\0.1&0.6\end{bmatrix}.
$$


```
N = sp.Matrix([[sp.Rational(9,10), sp.Rational(4,10)], [sp.Rational(1,10), sp.Rational(6,10)]])
sol_N = sp.solve([sp.Eq((N*sp.Matrix([p1,p2]))[0], p1), sp.Eq(p1+p2, 1)], [p1,p2])
sol_N
```

**Answer.** The steady state is $(0.8,0.2)^T$.

## 7. Power method

The power method approximates a dominant eigenvector by repeated multiplication and normalization.


```
A = np.array([[4, 1], [2, 3]], dtype=float)
x = np.array([1.0, 0.5])
for k in range(10):
    x = A @ x
    x = x / np.linalg.norm(x)
    rayleigh = (x @ A @ x) / (x @ x)
    print(k+1, x, rayleigh)
```

The Rayleigh quotient approaches the dominant eigenvalue $5$, and the vector approaches the dominant eigendirection $(1,1)^T$.
